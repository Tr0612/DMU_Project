```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import scienceplots
```

```python
nsteps_mt1 = [9e3, 3e4, 9e4, 2e5]
nsteps_ml1 = [19000, 37930, 96735, 201514]
mt1_success = np.array([0, 1, 2, 6])
no_train_success = np.array([0, 0, 0, 0])
train_success = np.array([0, 5, 6, 10])
```

```python
plt.style.use(['science', 'ieee'])
plt.set_cmap('viridis')
plt.plot(nsteps_ml1, train_success, label="window-close ML1 (With Fine-Tuning)", color=plt.cm.viridis(0))
plt.plot(nsteps_mt1, mt1_success, label="window-close MT1", color=plt.cm.viridis(0.4))
plt.plot(nsteps_mt1, no_train_success, label="window-close ML1 (No Fine-Tuning)", color=plt.cm.viridis(0.8))
plt.xlabel('Total Number of Environment Steps Including Fine-Tuning')
plt.ylabel('Number of Test Successes, Out of 10')
# plt.xscale('log')
plt.legend(bbox_to_anchor=(0.5, -0.5), loc='lower center')
plt.savefig('ml1-curve')
plt.show()
```
